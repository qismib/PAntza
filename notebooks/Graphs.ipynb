```python
import matplotlib.pyplot as plt

# Dati
x = [-0.853, -1.0863678799844634, -1.099465917941805, -1.075, -1.096, -1.006, -1.100, -1.137]
y = [4, 2, 3, 4, 5, 6, 7, 4]
annotazioni = ['Circuito :Gate \nAnsatz :Rand', 'Circuito: gate \nAnsatz: HEA', 'Circuito: pulse \nAnsatz: HEA', 
               'Circuito: pulse \nAnsatz: HEA', 'Circuito: pulse \nAnsatz: HEA', 'Circuito: gate \nAnsatz: RY-CNOT', 
               'Circuito: gate \nAnsatz: RY-CNOT', 'teorico']
colori = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'brown', 'yellow']
specifiche = [
    'backend: ibm_jakarta \nDimensione Ansatz[ns]: 682.7', 'backend: ibm_lagos \nDimensione Ansatz[ns]: 380', 
    'backend: ibm_lagos \nDimensione Ansatz[ns]: 50', 'backend: ibm_nazca \nDimensione Ansatz[ns]: 128 \nTempo esecuzione[s]: 40.37', 
    'backend: ibm_osaka \nDimensione Ansatz[ns]: 128 \nTempo esecuzione[s]: 58.11', 'backend: ibmq_manila \nDimensione Ansatz[ns]: -- \nTempo esecuzione[s]: 334', 
    'backend: ibmq_manila \nDimensione Ansatz[ns]: --  \nTempo esecuzione[s]: 13297', ''
]
annotazioni_specifiche = [f'{a}\n{s}' for a, s in zip(annotazioni, specifiche)]

# Crea il grafico a dispersione
fig, ax = plt.subplots(figsize=(10, 6))

# Lista di handle per la legenda
legend_elements = []

for i in range(len(x)):
    if 'ibm_nazca' in specifiche[i] or 'ibm_osaka' in specifiche[i]:
        # Evidenzia i punti con marker giallo e bordo nero
        ax.scatter(x[i], y[i], color=colori[i], edgecolor='black', linewidth=2, s=100, label=annotazioni_specifiche[i])
        # Aggiungi l'elemento alla legenda se non esiste gi
        if annotazioni_specifiche[i] not in [element.get_label() for element in legend_elements]:
            legend_elements.append(
                plt.Line2D([0], [0], marker='o', color=colori[i], markerfacecolor=colori[i], markeredgecolor='black',
                           markeredgewidth=2, markersize=10, label=annotazioni_specifiche[i])
            )
    else:
        # Altri punti
        ax.scatter(x[i], y[i], color=colori[i], edgecolor='black', linewidth=0.5, s=100)
        # Aggiungi l'elemento alla legenda se non esiste gi
        if annotazioni_specifiche[i] not in [element.get_label() for element in legend_elements]:
            legend_elements.append(
                plt.Line2D([0], [0], marker='o', color=colori[i], markerfacecolor=colori[i], markeredgecolor='none',
                           markeredgewidth=2, markersize=10, label=annotazioni_specifiche[i])
            )

# Annotazioni con frecce per i punti specifici
annotazioni_frecce = {
    'ibm_nazca': r'-1.075 $E_H$',
    'ibm_osaka': r'-1.096 $E_H$',
    'teorico': r'-1.137 $E_H$'
}

for i, (x_val, y_val) in enumerate(zip(x, y)):
    if 'ibm_nazca' in specifiche[i]:
        ax.annotate(annotazioni_frecce['ibm_nazca'], xy=(x_val, y_val), xytext=(x_val + 0.05, y_val + 0.5),
                    arrowprops=dict(facecolor='black', arrowstyle='->', shrinkA=0, shrinkB=0))
    elif 'ibm_osaka' in specifiche[i]:
        ax.annotate(annotazioni_frecce['ibm_osaka'], xy=(x_val, y_val), xytext=(x_val + 0.05, y_val + 0.5),
                    arrowprops=dict(facecolor='black', arrowstyle='->', shrinkA=0, shrinkB=0))
    elif 'teorico' in annotazioni[i]:
        ax.annotate(annotazioni_frecce['teorico'], xy=(x_val, y_val), xytext=(x_val + 0.05, y_val + 0.5),
                    arrowprops=dict(facecolor='black', arrowstyle='->', shrinkA=0, shrinkB=0))

# Rimuovi l'asse delle y
ax.axes.get_yaxis().set_visible(False)
ax.invert_xaxis()

# Aggiungi etichetta all'asse delle x
ax.set_xlabel('Energia (Hartree)')

# Aggiungi la legenda
legend = ax.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)

# Aggiungi riquadri attorno ai gruppi
for text in legend.get_texts():
    text.set_bbox(dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.title(r'$H_2$')
plt.tight_layout()  # Per fare spazio alla legenda fuori dal grafico
output_path = 'fin_h2.pdf'
plt.savefig(output_path)
plt.show()
```
