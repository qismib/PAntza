```python
import matplotlib.pyplot as plt
import numpy as np

# Leggi i dati dal file
file_path = 'EnergyVSiterH2.txt'
data = np.loadtxt(file_path)

# Estrai colonne x ed y
x = data[:, 0]
y = data[:, 1]

# Crea il plot
plt.plot(x, y, label='Energy vs Iteration')
plt.xlabel('Iteration')
plt.ylabel('Energy')
plt.title('Energy vs Iteration Plot')
plt.legend()

# Salva il plot come immagine PNG
output_path = 'EnergyVSiterH2.png'
plt.savefig(output_path)

# Mostra il plot
plt.show()
```
